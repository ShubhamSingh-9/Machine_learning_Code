```python
ipl
```

```python
plt.plot(ipl['index'],ipl['V Kohli'])
plt.plot(ipl['index'],ipl['RG Sharma'])

plt.xlabel('Season')
plt.ylabel('Runs')
plt.title('Rohot vs Kohli Career')
```

```python
# colors(hex) and line(width and style) and marker(size)
#just give hex code(color code)
plt.plot(ipl['index'],ipl['V Kohli'],color= 'blue',linestyle= 'dashed', linewidth= 3)
plt.plot(ipl['index'],ipl['RG Sharma'],color='black',linestyle= 'dashdot', linewidth= 2)

plt.xlabel('Season')
plt.ylabel('Runs')
plt.title('Rohot vs Kohli Career');
```

```python
# colors(hex) and line(width and style) and marker(size)
#just give hex code(color code)
plt.plot(ipl['index'],ipl['V Kohli'],color= 'blue',linestyle= 'dashed', linewidth= 3,marker= 'D')
plt.plot(ipl['index'],ipl['RG Sharma'],color='black',linestyle= 'dashdot', linewidth= 2, marker= '+')

plt.xlabel('Season')
plt.ylabel('Runs')
plt.title('Rohot vs Kohli Career');
```

```python
# legend-> location naming the line

# colors(hex) and line(width and style) and marker(size)
#just give hex code(color code)
plt.plot(ipl['index'],ipl['V Kohli'],color= 'blue',linestyle= 'dashed', linewidth= 3,marker= 'D', label= 'Virat')
plt.plot(ipl['index'],ipl['RG Sharma'],color='black',linestyle= 'dashdot', linewidth= 2, marker= '+', label= 'Rahul')

plt.xlabel('Season')
plt.ylabel('Runs')
plt.title('Rohot vs Kohli Career')
plt.legend()  #legend bcs need to get he label printed and default parameter is loc=best so take the best postition
plt.grid()
plt.show();  #need to add this as if plottng the graph in some other application like pycharm it will help to plot the graph
```

```python
# limiting Axes

price =[48000,54000,57000,49000,47000,45000,4500000]
year= [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year, price);
```

```python
# To get rid of the outlier we apply limit on the axis and its as per our need 
price =[48000,54000,57000,49000,47000,45000,4500000]
year= [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year, price)
plt.ylim(0,100000)
plt.xlim(2017,2019);
```

```python
x= np.linspace(-10,10,50)
y= 10*x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y);
```

```python
df= pd.read_csv('batter.csv')  
plt.scatter(df['avg'],df['strike_rate'],color= 'red',marker= '+') 
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Avg')
plt.ylabel('SR')
```

```python
# Size

df1= sns.load_dataset('tips')
df1.head()

plt.scatter(df1['total_bill'],df1['tip'],s= df1['size']*20)
#s take size and as more people big the dot size
```

```python
#other way to make it work the size, color parameter cant be used in plt.plot and its not suggested to use  because its faster

plt.plot(df1['total_bill'],df1['tip'],'o')
```

# Pie Chart

1. Univarient and Bivarient
2. Categorical vs Numerical
3. To find contribution on a standard scale

```python
data= [80,90,100,80,79]
subject= ['eng','sci','ml','sst','hindi']
plt.pie(data,labels= subject);
```

```python
df4= pd.read_csv('gayle-175.csv')
df4.head()
```

```python
plt.pie(df4['batsman_runs'],labels= df4['batsman'],autopct= '%0.1f%%')
plt.show()
#autopct means auto percentage
```

```python
#specify color 
plt.pie(df4['batsman_runs'],labels= df4['batsman'],autopct= '%0.1f%%',colors=['blue','green','yellow','black','red','violet'])
plt.show()
```

```python
#show part of data

plt.pie(df4['batsman_runs'],labels= df4['batsman'],autopct= '%0.1f%%',colors=['blue','green','yellow','black','red','violet'],explode= [0.2,0,0,0,0,0],shadow= True)
plt.show()
```
