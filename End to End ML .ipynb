# In this notebook will build ML using End to End method.

```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Getting the data

df=pd.read_csv('placement.csv')
df
```

## Step 1: Preprocess + EDA + Feature Selection

```python
# Getting infromation from the data

df.info()
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'],c=df['placed']);
```

## Step 2: Extract input and output cols

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

## Step 3: Scale the value

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

## Step 4: Train Test Split

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state =2)
```

## Step 5: Train the Model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
y_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```
