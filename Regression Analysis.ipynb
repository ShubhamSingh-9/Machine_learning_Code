```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x= 10*np.random.rand(50)
y= 3 * x - 8 + np.random.randn(50)*4
```

```python
print(x.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
x= x.reshape(-1,1)
lr= LinearRegression()
lr.fit(x,y)
```

```python
y_pred= lr.predict(x)
```

```python
plt.scatter(x, y, label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot with increased variability and regression lines")
```

```python
x_line = np.linspace(0, 10, 100)
y_actual = 2 * x_line - 5
plt.plot(x_line, y_actual, 'r', label="Population line (m=3, b=-8)")
plt.legend()
plt.show
```

```python
y_estimated = lr.coef_[0] * x_line + lr.intercept_
plt.plot(x_line, y_estimated, 'g', label=f"Estimated line (m={lr.coef_[0]:.2f}, b={lr.intercept_:.2f})")
plt.legend()
plt.show()
```
