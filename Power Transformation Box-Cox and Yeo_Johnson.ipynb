# Box-Cox is trabnsfornation that is applied to data to make it normally distribution the square and square root are type of Box Plot we need to apply and see at which lambda value we get a good normally distributed data. 

# Yeo-Johnson is applied to data when the value 0 or -ve earlier all transformation we dont carry out with 0 and -ve value.

# Box-Cox value lambda is calcualyed using Maximum Likelyhood as its ML model

# Yeo-Johnson is by deafult value and it give better value than Box-Cox 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df= pd.read_csv('concrete_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
```

```python
df.describe()  #if value is -ve or 0 cannt apply Box_cox
```

```python
X= df.iloc[:,0:-1]
y= df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size= 0.2, random_state= 20)
```

```python
lr= LinearRegression()
lr.fit(X_train,y_train)
y_pred= lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
lr1= LinearRegression()
np.mean(cross_val_score(lr,X_train,y_train,scoring= 'r2'))
```

```python
for col in X_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    #sns.histplot(X_train[col],kde=True)
    plt.title(col)
    
    plt.subplot(122)
    stats.probplot(X_train[col], dist= 'norm', plot= plt)
    plt.title(col)
    
    plt.show()
```

```python
pt= PowerTransformer(method='box-cox')

X_train_transformed= pt.fit_transform(X_train+0.001)
X_test_transformed= pt.transform(X_test+0.001)

pd.DataFrame({'cols':X_train.columns,'box-cox_lambdas': pt.lambdas_})
```

```python
lr2= LinearRegression()

lr2.fit(X_train_transformed,y_train)
y_pred= lr2.predict(X_test_transformed)
r2_score(y_test,y_pred)
```

```python
lr3= LinearRegression()

np.mean(cross_val_score(lr3,X_train_transformed,y_train,scoring= 'r2'))
```

```python
X_train_transformed= pd.DataFrame(X_train_transformed)
X_train_transformed.columns= X_train.columns
```

```python
for col in X_train_transformed.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.histplot(X_train[col], kde= True)
    #sns.histplot(X_train[col],kde=True)
    plt.title('before')

    plt.subplot(122)
    sns.histplot(X_train_transformed[col],kde= True)
    #stats.probplot(X_train_transformed[col], dist= 'norm', plot= plt)
    plt.title('After')
    
    plt.show()
```

```python
ye= PowerTransformer()

X_train_yeo= ye.fit_transform(X_train)
X_test_yeo= ye.transform(X_test)

lr4= LinearRegression()

lr4.fit(X_train_yeo,y_train)
y_pred= lr4.predict(X_test_yeo)
print(r2_score(y_test,y_pred))

pd.DataFrame({'cols': X_train.columns, 'Yeo_Johnson_lambdas': ye.lambdas_})
X_train_yeo= pd.DataFrame(X_train_yeo,columns=X_train.columns)
```

```python
np.mean(cross_val_score(lr4,X_train_yeo,y_train,scoring= 'r2'))
```

```python
for col in X_train_yeo.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.histplot(X_train_transformed[col], kde= True)
    #sns.histplot(X_train[col],kde=True)
    plt.title('Box-Cox')

    plt.subplot(122)
    sns.histplot(X_train_yeo[col],kde= True)
    #stats.probplot(X_train_transformed[col], dist= 'norm', plot= plt)
    plt.title('Yeo-Johnson')
    
    plt.show()
```
