```python
import numpy as np
import pandas as pd

df= pd.read_csv('heart.csv')
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size= 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
cf1= LogisticRegression()
cf2= DecisionTreeClassifier()
```

```python
cf1.fit(X_train,y_train)
cf2.fit(X_train,y_train)

ypred1= cf1.predict(X_test)
ypred2= cf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

print('Accuracy of Logistic Regression', accuracy_score(y_test,ypred1))
print('Accuracy of Decison Tree Classifier', accuracy_score(y_test,ypred2))
```

```python
confusion_matrix(y_test,ypred1)
```

```python
confusion_matrix(y_test,ypred2)
```
